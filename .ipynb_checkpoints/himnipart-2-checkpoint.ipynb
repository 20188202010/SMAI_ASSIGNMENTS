```python
import pandas as pd
import numpy as np
import sys
import matplotlib
import matplotlib.pyplot as plt
from pylab import *
```

```python
class Node:
    def __init__(self, name, pos,neg):
        self.name = name
        self.positive = pos
        self.negative=neg
        self.children = {}
        self.isLeaf=False
        self.split=-1
        if(pos>neg):
            self.result=1
        else:
            self.result=0
    def add_child(self, key,value):
        self.children[key]=value
```

```python
# filename=raw_input('Enter file name:')
filename='train.csv'
original_dataset=pd.read_csv(filename)
#split data
df=pd.DataFrame(original_dataset)
train_dataset,validate_dataset=np.split(df,[int(0.8*len(df))])
# train_categorical=train_dataset.filter(['Work_accident','promotion_last_5years','sales','salary','left'],axis=1)
# train_categorical=train_dataset.filter(['temperature','outlook','humidity','windy','play'],axis=1)
# train_categorical=df.filter(['temperature','outlook','humidity','windy','play'],axis=1)
# validate_categorical=validate_dataset.filter(['Work_accident','promotion_last_5years','sales','salary','left'],axis=1)
# validate_categorical=validate_dataset.filter(['temperature','outlook','humidity','windy','play'],axis=1)
print train_dataset
print validate_dataset
continuous_attr=['satisfaction_level','last_evaluation','number_project','average_montly_hours','time_spend_company']
# categorical_attr=['Work_accident','promotion_last_5years','sales','salary','left']
```

```python
def cal_entropy(df,col_name):
    ent=0
    unique_values=df[col_name].unique()
    total_len=len(df[col_name])
    for i in unique_values:
        tmp=float(df[col_name].value_counts()[i])/total_len
#         print tmp
        if(tmp!=0):
            ent+=-tmp*np.log2(tmp)
    return ent
q=cal_entropy(train_dataset,'left')
# q=cal_entropy(train_categorical,'play')
print q
    
```

```python
def cal_entropy_attribute(df,col_name,result_col):
    values=df[result_col].unique()
    attributes=df[col_name].unique()
    total_entropy=0
    total_len=len(df[col_name])
    for x in attributes:
        tmp_entropy=0
        attribute_tmp_dataset=df.loc[df[col_name]==x]
        attr_len=len(attribute_tmp_dataset[col_name])
        for i in values:
            if  attribute_tmp_dataset[attribute_tmp_dataset[result_col]==i].empty:
                continue
            tmp=float(attribute_tmp_dataset[result_col].value_counts()[i])/attr_len
            if(tmp!=0):
                tmp_entropy+=-tmp*np.log2(tmp)
        total_entropy+=(float(attr_len)/total_len)*tmp_entropy
    return total_entropy
# cal_entropy_attribute(train_categorical,'promotion_last_5years','left')
    
```

```python
def cal_numerical_entropy_attribute(df,col_name):
    attributes=df[col_name].unique()
    values=df['left'].unique()
    dict_entropy={}
    total_len=len(df[col_name])
#     print "length: ",total_len
    for x in attributes:
        attr_le_dataset=df.loc[df[col_name]<=x]
        attr_gt_dataset=df.loc[df[col_name]>x]
#         print "x: ",x,
#         print attr_le_dataset
        attr_le=len(attr_le_dataset[col_name])
        attr_gt=len(attr_gt_dataset[col_name])
        le_entropy=cal_entropy(attr_le_dataset,'left')
        gt_entropy=cal_entropy(attr_gt_dataset,'left')
        total_entropy=(float(attr_le)/total_len)*le_entropy
        total_entropy+=(float(attr_gt)/total_len)*gt_entropy
#         print "left: ",le_entropy," right: ",gt_entropy," total_entropy: ",total_entropy
        dict_entropy[x]=total_entropy
    min_key=min(dict_entropy, key=dict_entropy.get)
#     print len(dict_entropy)
#     for i in dict_entropy.keys():
#         print "key:",i," value: ",dict_entropy[i]
    return min_key,dict_entropy[min_key]

cal_numerical_entropy_attribute(train_dataset,'satisfaction_level')
        
```

```python
def max_gain(df):
    cols=list(df.columns.values)
    gain_max=0
    key=cols[0]
    entropy=cal_entropy(df,'left')
    split_num=-1
    for c in cols:
        if(c!='left'):
            tmp=0
            if c in continuous_attr:
                split,ent_tmp=cal_numerical_entropy_attribute(df,c)
                tmp=entropy-ent_tmp
#                 print cal_numerical_entropy_attribute(df,c)
#                 print "split:",split
            else:
                tmp=entropy-cal_entropy_attribute(df,c,'left')
#                 print "attr: ",cal_entropy_attribute(df,c,'left'),
            if(tmp>gain_max):
                gain_max=tmp
                key=c
                if c in continuous_attr:
                    split_max=split
                else:
                    split_max=-1
#             print "c: ",c," key: ",key," gain: ",tmp," split_max: ",split_max
    return key,split_max
max_gain(train_dataset)
```

```python
def cal_impurity(df,col_name,val):
    if df[df[col_name]==val].empty:
        print 'empty'
        return 0
#     tmp=df['play'].value_counts()[val]
    tmp=df[col_name].value_counts()[val]
    return tmp
# cal_impurity(train_categorical,'left',1)
```

```python
def build_tree(df):
    pos=cal_impurity(df,'left',1)
    neg=cal_impurity(df,'left',0)
#     print "pos, neg:",pos,neg
    if(pos==0):
        root=Node(0,0,neg)
        root.isLeaf=True
        return root
    if(neg==0):
        root=Node(1,pos,0)
        root.isLeaf=True
        return root

    
    node,split=max_gain(df)
    root=Node(node,pos,neg)
#     print "node: ",node
    if node in continuous_attr:
#         print "split",split
        root.split=split
        le_df=df[df[node]<=split]
#         print le_df
        gt_df=df[df[node]>split]
#         print "left_len",len(le_df),"gttree",len(gt_df)
    
        le_tree=build_tree(le_df)
        gt_tree=build_tree(gt_df)
        root.add_child("<="+str(split),le_tree)
        root.add_child(">"+str(split),gt_tree)
        
    else:
        attributes=df[node].unique()
        for value in attributes:
            print "value: ", value
            df_tmp=df[df[node]==value]
    #         print "df_temp"
#             df_tmp=df_tmp.drop(node, 1)
    #         print "dropped"
    #         print df_tmp
            tree=build_tree(df_tmp)
#             print "tree name:",tree.name,value
            root.add_child(value,tree)
#         print '-------------------------------------------------------'
  
    return root
ans=build_tree(train_dataset)
```

```python
def traverse(root):
    if root.isLeaf:
        print "root leaf:",root.name
        return
    
    print "Root: ",root.name
    
    for k,v in root.children.items():
        print "root: ",root.name, "key: ",k
        traverse(v)
traverse(ans)
```

```python
def prediction(row,root):
    if(root.isLeaf):
        return root.name
    col=root.name
    val=row[col]
    if col in continuous_attr:
        if(val<=root.split):
            key='<='+str(root.split)
            return prediction(row,root.children[key])
        else:
            key='>'+str(root.split)
            return prediction(row,root.children[key])
    else:
        if val in root.children.keys():
            return prediction(row,root.children[val])
    return root.result
    

def helper(df,root,col_name):
     df[col_name] = df.apply(prediction, axis=1, args=(root,) )
helper(validate_dataset,ans,'predict')
# validate_dataset['predict']
```

```python
def measures(df,col_name):
    TP=0
    FP=0    
    TN=0    
    FN=0
    for index,row in df.iterrows():
        if(row['left']==0 and row['left']==row[col_name]):
            TN+=1
        elif (row['left']==1 and row['left']==row[col_name]):
            TP+=1
        elif(row['left']==0 and row[col_name]==1):
            FP+=1
        elif (row['left']==1 and row[col_name]==0):
            FN+=1
    return TP,TN,FP,FN
#
tp,tn,fp,fn=measures(validate_dataset,'predict')
print len(validate_dataset)
print tp,tn,fp,fn
```

```python
def precision(tp,tn,fp,fn):
    return float(tp)/(tp+fp)

def recall(tp,tn,fp,fn):
    return float(tp)/(tp+fn)

def accuracy(tp,tn,fp,fn):
    return float(tp+tn)/(tp+fn+tn+fp)
    
def f1_score(tp,tn,fp,fn):
    return 2/((1/float(recall(tp,tn,fp,fn)))+(1/float(precision(tp,tn,fp,fn))))

print precision(tp,tn,fp,fn)
print recall(tp,tn,fp,fn)
print accuracy(tp,tn,fp,fn)
print f1_score(tp,tn,fp,fn)
```

```python
def cal_gini_index(df,col_name):
    gini=0
#     unique_values=df[col_name].unique()
    total_len=len(df[col_name])
#     for i in unique_values:
    tmp_df=df[df[col_name]==1]
    len_df=len(tmp_df)
    if(total_len!=0):
        q=float(len_df)/total_len
    else:
        q=0
#     tmp=float(df[col_name].value_counts()[1])/total_len
#         print tmp
    gini+=q*2*(1-q)
    return gini
cal_gini_index(train_dataset,'left')
```

```python
def cal_gini_attr(df,col_name,result_col):
#     values=df[result_col].unique()
    attributes=df[col_name].unique()
    total_gini=0
    total_len=len(df[col_name])
    for x in attributes:
        attribute_tmp_dataset=df[df[col_name]==x]
        attr_len=len(attribute_tmp_dataset[col_name])
        tmp_df=attribute_tmp_dataset[attribute_tmp_dataset[result_col]==1]
        len_tmp_df=len(tmp_df)
#         print "len_tmp_df",len_tmp_df," attrlen ",attr_len
        if attr_len!=0:
            q=float(len_tmp_df)/attr_len
        else:
            q=0
#         print q
        tmp_gini=2*q*(1-q)
        total_gini+=(float(attr_len)/total_len)*tmp_gini
    return total_gini
for colname in train_dataset:
    print colname,",",cal_gini_attr(train_dataset,colname,'left')
```

```python
def cal_numerical_gini_attr(df,col_name):
    attributes=df[col_name].unique()
#     values=df['left'].unique()
    dict_gini={}
    total_len=len(df[col_name])
#     print "length: ",total_len
    for x in attributes:
        attr_le_dataset=df.loc[df[col_name]<=x]
        attr_gt_dataset=df.loc[df[col_name]>x]
#         print "x: ",x,
#         print attr_le_dataset
        attr_le=len(attr_le_dataset[col_name])
        attr_gt=len(attr_gt_dataset[col_name])
        le_gini=cal_gini_index(attr_le_dataset,'left')
        gt_gini=cal_gini_index(attr_gt_dataset,'left')
        total_gini=(float(attr_le)/total_len)*le_gini
        total_gini+=(float(attr_gt)/total_len)*gt_gini
#         print "left: ",le_entropy," right: ",gt_entropy," total_entropy: ",total_entropy
        dict_gini[x]=total_gini
    min_key=min(dict_gini, key=dict_gini.get)
#     print len(dict_entropy)
#     for i in dict_entropy.keys():
#         print "key:",i," value: ",dict_entropy[i]
    return min_key,dict_gini[min_key]

cal_numerical_gini_attr(train_dataset,'satisfaction_level')
        
```

```python
def max_gini_gain(df):
    cols=list(df.columns.values)
    gain_max=0
    key=cols[0]
    gini=cal_gini_index(df,'left')
    split_max=-1
    for c in cols:
        if(c!='left'):
            tmp=0
            if c in continuous_attr:
                split,gini_tmp=cal_numerical_gini_attr(df,c)
                tmp=gini-gini_tmp
#                 print cal_numerical_entropy_attribute(df,c)
#                 print "split:",split
            else:
                tmp=gini-cal_gini_attr(df,c,'left')
#                 print "attr: ",cal_entropy_attribute(df,c,'left'),
            if(tmp>gain_max):
                gain_max=tmp
                key=c
                if c in continuous_attr:
                    split_max=split
                else:
                    split_max=-1
#             print "c: ",c," key: ",key," gain: ",tmp," split_max: ",split_max
    print " gain_max : ",gain_max
    return key,split_max
max_gini_gain(train_dataset)
```

```python
def build_tree_gini(df):
    pos=cal_impurity(df,'left',1)
    neg=cal_impurity(df,'left',0)
    if(pos==0):
        root=Node(0,0,neg)
        root.isLeaf=True
        return root
    if(neg==0):
        root=Node(1,pos,0)
        root.isLeaf=True
        return root
    node,split=max_gini_gain(df)
    root=Node(node,pos,neg)
    print "node: ",node,
    if node in continuous_attr:
        root.split=split
        le_df=df[df[node]<=split]
        gt_df=df[df[node]>split]
        le_tree=build_tree_gini(le_df)
        gt_tree=build_tree_gini(gt_df)
        root.add_child("<="+str(split),le_tree)
        root.add_child(">"+str(split),gt_tree)
        
    else:
        attributes=df[node].unique()
        for value in attributes:
            print "value: ", value
            df_tmp=df[df[node]==value]
            tree=build_tree_gini(df_tmp)
            root.add_child(value,tree)
  
    return root
ans_gini=build_tree_gini(train_dataset)
```

```python
helper(validate_dataset,ans_gini,'predict_gini')
# helper(validate_dataset)
tp,tn,fp,fn=measures(validate_dataset,'predict_gini')
print precision(tp,tn,fp,fn)
print recall(tp,tn,fp,fn)
print accuracy(tp,tn,fp,fn)
print f1_score(tp,tn,fp,fn)
```

## Misclassification

```python
def cal_misclassification(df,col_name):
    misclas=0
    total_len=len(df[col_name])
    tmp_df=df[df[col_name]==1]
    len_df=len(tmp_df)
    if(total_len!=0):
        q=float(len_df)/total_len
    else:
        q=0
    misclas=min(q,1-q)
    return misclas
cal_misclassification(train_dataset,'left')
```

```python
def cal_misclass_attr(df,col_name,result_col):
    attributes=df[col_name].unique()
    total_mc=0
    total_len=len(df[col_name])
    for x in attributes:
        attribute_tmp_dataset=df[df[col_name]==x]
        attr_len=len(attribute_tmp_dataset[col_name])
        tmp_df=attribute_tmp_dataset[attribute_tmp_dataset[result_col]==1]
        len_tmp_df=len(tmp_df)
        if attr_len!=0:
            q=float(len_tmp_df)/attr_len
        else:
            q=0
        tmp_mc=min(q,1-q)
        total_mc+=(float(attr_len)/total_len)*tmp_mc
    return total_mc
print cal_misclass_attr(train_dataset,'salary','left')
# print cal_misclass_attr(train_dataset,'sales','left')
```

```python
def cal_misclass_numerical(df,col_name):
    attributes=df[col_name].unique()
#     values=df['left'].unique()
    dict_mc={}
    total_len=len(df[col_name])
#     print "length: ",total_len
    for x in attributes:
        attr_le_dataset=df.loc[df[col_name]<=x]
        attr_gt_dataset=df.loc[df[col_name]>x]
#         print "x: ",x,
#         print attr_le_dataset
        attr_le=len(attr_le_dataset[col_name])
        attr_gt=len(attr_gt_dataset[col_name])
        le_mc=cal_misclassification(attr_le_dataset,'left')
        gt_mc=cal_misclassification(attr_gt_dataset,'left')
        total_mc=(float(attr_le)/total_len)*le_mc
        total_mc+=(float(attr_gt)/total_len)*gt_mc
#         print "left: ",le_mc," right: ",gt_mc," total_entropy: ",total_mc
#         print " ln ",float(attr_le)/total_len," gt ",float(attr_gt)/total_len
        dict_mc[x]=total_mc
    min_key=min(dict_mc, key=dict_mc.get)
#     print len(dict_entropy)
#     for i in dict_entropy.keys():
#         print "key:",i," value: ",dict_entropy[i]
    return min_key,dict_mc[min_key]

cal_misclass_numerical(train_dataset,'satisfaction_level')
        
```

```python
def max_misclas_gain(df):
    cols=list(df.columns.values)
    gain_max=0
    key=cols[0]
    misclas=cal_misclassification(df,'left')
#     print "misclas: ",misclas,
    split_max=-1
    for c in cols:
        if(c!='left'):
            tmp=0
            if c in continuous_attr:
                split,mc_tmp=cal_misclass_numerical(df,c)
                tmp=misclas-mc_tmp
#                 print " numeric:",cal_misclass_numerical(df,c),
#                 print "split:",split
            else:
                tmp=misclas-cal_misclass_attr(df,c,'left')
#                 print" cat: ",cal_misclass_attr(df,c,'left'),
#                 print "attr: ",cal_entropy_attribute(df,c,'left'),
            if(tmp>gain_max):
                gain_max=tmp
                key=c
                if c in continuous_attr:
                    split_max=split
                else:
                    split_max=-1
#             print "c: ",c," key: ",key," gain: ",tmp," split_max: ",split_max
    print " gain_max: ",gain_max
    return key,split_max,gain_max
max_misclas_gain(train_dataset)
```

```python
def build_tree_misclas(df):
#     print "###############################"
    pos=cal_impurity(df,'left',1)
    neg=cal_impurity(df,'left',0)
#     print "pos: ",pos," neg: ",neg,
#     if(pos==0):
#         root=Node(0,0,neg)
#         root.isLeaf=True
#         print "return pos"
#         return root
#     if(neg==0):
#         root=Node(1,pos,0)
#         root.isLeaf=True
#         print "return neg" 
#         return root
    
    node,split,gain=max_misclas_gain(df)
    if(gain==0):
        if(pos>neg):
            root=Node(1,pos,neg)
            root.isLeaf=True
            return root
        else:
            root=Node(0,pos,neg)
            root.isLeaf=True
            return root
    root=Node(node,pos,neg)
    print "node: ",node,
    if node in continuous_attr:
        root.split=split
#         print "**if**",
        print "split: ",split,
        le_df=df[df[node]<=split]
        gt_df=df[df[node]>split]
            
        print "ledf: ",len(le_df)," gt_df: ",len(gt_df)
        le_tree=build_tree_misclas(le_df)
        gt_tree=build_tree_misclas(gt_df)
#         print "###############################"
#         print "left return"
#         print "right return"
        
        root.add_child(">"+str(split),gt_tree)
        root.add_child("<="+str(split),le_tree)
#         print "end"
    else:
#         print "**else**",
        attributes=df[node].unique()
        for value in attributes:
#             print "value: ", value
            df_tmp=df[df[node]==value]
#             df_tmp=df_tmp.drop(node,1)
#             print "###############################"
            tree=build_tree_misclas(df_tmp)
            root.add_child(value,tree)
#     print "--------------------------------------------------------------------"
#     print "root"
    return root
ans_misclas=build_tree_misclas(train_dataset)
```
