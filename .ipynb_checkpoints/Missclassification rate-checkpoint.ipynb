```python
def valid_or_not(inst,tree):
    global cat_list
    global num_list
    #This function is used to predict for any input variable 
    
    #Recursively we go through the tree that we built earlier
#     print(tree)
    for nodes in tree.keys():        
        value = inst[nodes]
        if nodes in cat_list:
            if value in list((tree[nodes]).keys()):
                tree = tree[nodes][value]
            else:
                zeros = 0
                ones = 0
                for i in tree[nodes].keys():
                    if tree[nodes][i] == 0:
                        zeros += 1
                    elif tree[nodes][i] == 1:
                        ones += 1
                if zeros > ones:
                    return 0
                else:
                    return 1
                  
        elif nodes in num_list:
            first_key = list(tree[nodes].keys())[0]
            if value <= first_key :
                tree = tree[nodes][first_key]
            else:
                second_key = list(tree[nodes].keys())[1]
                tree = tree[nodes][second_key]
        prediction = 0

        if type(tree) is dict:
            prediction = valid_or_not(inst, tree)
        else:
            prediction = tree
            break;
        
    return prediction
   
```

```python
def validation(validate, tree):
    global true_positive
    global true_negative
    global false_positive
    global false_negative
    for index, row in validate.iterrows():
        x = valid_or_not(row,tree)
        if x == row["left"]:
            if x == 1:
                true_positive += 1
            elif x == 0:
                true_negative += 1
        
        else:
            if x == 0:
                false_negative += 1
            elif x == 1:
                false_positive += 1
```

```python
def find_entropy(df):
    Class = "left"   #To make the code generic, changing target variable class name
#     Class = "play"   #To make the code generic, changing target variable class name
    entropy = sys.maxsize
    values = df[Class].unique()
    for value in values:
        fraction = df[Class].value_counts()[value]/float(len(df[Class]))
        entropy = min(fraction, entropy)
 
    return entropy
```

```python
import sys
def find_entropy_attribute(df,attribute):
  Class = "left"   #To make the code generic, changing target variable class name
#   Class = "play"   #To make the code generic, changing target variable class name
  target_variables = df[Class].unique()  #This gives all 'Yes' and 'No'
  variables = df[attribute].unique()    #This gives different features in that attribute (like 'Hot','Cold' in Temperature)
  entropy2 = 0
  temp = {}
  for variable in variables:
      entropy = sys.maxsize
      for target_variable in target_variables:
          num = len(df[attribute][df[attribute]==variable][df[Class] ==target_variable])
          den = len(df[attribute][df[attribute]==variable])
          fraction = num/(den+eps)
          if num == 0 or num == den:
              entropy += 0.0
          else :
              entropy = min(fraction, entropy)
      fraction2 = den/len(df)
      entropy2 += -fraction2*entropy
  return abs(entropy2)
```

```python
def find_entropy_attribute_num(df, attribute):
    Class = "left"   #To make the code generic, changing target variable class name
    target_variables = df[Class].unique()  #This gives '1' and '0'
    variables = df[attribute].unique()    #This gives different features in that attribute (like 'Hot','Cold' in Temperature)
    temp = 0
    mini = 0
    i = 0
    df = pd.DataFrame(df, columns=[attribute,'left'])
    df.sort_values(by=[attribute])
    for val in variables:
        
        entropy_le = sys.maxsize
        entropy_gt = sys.maxsize
        # for target_variable in target_variables:
        target_variable = target_variables[0]
        total = len(df[attribute][df[Class] == target_variable]) #suppose this is for 1, then we got total number 
                                                                     # of 1's here
        total_length_col = len(df)
        # for 1's above and on the split line-----------
        num_le = len(df[attribute][df[attribute]<=val][df[Class] == target_variable])
        den_le = len(df[attribute][df[attribute]<=val])
        # for 1's below the split line---------------
        num_gt = total - num_le
        den_gt = total_length_col - den_le
        
        
        fraction_le = num_le/(den_le + eps)
        fraction_gt = num_gt/(den_gt + eps)

        entropy_le = min(fraction_le, entropy_le)
        entropy_gt = min(fraction_gt, entropy_gt)
        
        # now following same stpes for 0 also-----------
        num_le = den_le - num_le # number of 0's obtained from no of one's
        num_gt = total_length_col - total - num_le
        
        fraction_le0 = num_le/(den_le + eps)
        fraction_gt0 = num_gt/(den_gt + eps)

        entropy_le = min(fraction_le0, entropy_le)
        entropy_gt = min(fraction_gt0, entropy_gt)
        
# ---------------------------------------------------------------
        
        fraction2_le = den_le/(total_length_col + eps)
        entropy2_le = fraction2_le*entropy_le
        fraction2_gt = den_gt/(total_length_col + eps)
        entropy2_gt = fraction2_gt*entropy_gt
        if i == 0:
            mini = entropy2_le + entropy2_gt
            temp = val
        elif mini > entropy2_le + entropy2_gt:
            
            mini  = entropy2_le + entropy2_gt
            temp = val
        i = i + 1
    print("ATTRIBUTE IS :",attribute)
    print("TEMP IS : ",temp)
    return temp, mini
```

```python
def find_best_gain_cat(train, cols):
    weighted_sum = find_entropy_attribute(train,cols)
    print(" WEIGHTED SUM FOR CATAGORICAL DATA : ",weighted_sum)
    x = find_entropy(train) - weighted_sum
    return x
 
```

```python
def find_best_gain(train):
    
    global cat_list
    global num_list
    
    all_cols = list(train)
    list_of_gains = []
    i = 0
    spl = 0
    maximum = 0
    for cols in all_cols:
        if cols != "left":
            #         if cols != "play" :
            if cols in cat_list:
               
                if i is not 0:
                    z = find_best_gain_cat(train, cols)
                    if maximum < z:
                        maximum = z
                        list_of_gains.append(maximum)
                        att = cols
                else:
                    maximum = find_best_gain_cat(train, cols)
                    list_of_gains.append(maximum)
                    att = cols
            elif cols in num_list:
                split, weighted_sum = find_entropy_attribute_num(train,cols)

                print(" WEIGHTED SUM : ",weighted_sum)
                y = find_entropy(train) - weighted_sum
                list_of_gains.append(y)
                if i == 0:
                    maximum = y
                    att = cols
                    spl = split
                elif maximum < y:
                    maximum = y
                    att = cols
                    spl = split

                print(cols,list_of_gains[i])
            i += 1

    if(len(set(list_of_gains))==1):
        return "done", "done"
    print("list_of_gains : ",list_of_gains)
    return att, spl
 
```

```python
def get_subtable(df, node, value):
    return df[df[node] == value].reset_index(drop=True)
```

```python
def get_num_subtable(df, node, value):
    return df[df[node] <= value].reset_index(drop=True), df[df[node] > value].reset_index(drop=True)
```

```python
def buildTree(df,height,tree=None): 
    
#     print("DATFRAME INSIDE BUILDTREE")
#     print("--------------------------")
#     print(df)
#     print("--------------------------")

#     ipdb.set_trace()

    Class = "left"
    clValue,counts = np.unique(df['left'],return_counts=True)
    
    node, split = find_best_gain(df) # best gain node
    if node == "done" :
        return clValue[np.argmax(counts)]
    print(" BEST GAIN NODE : ",node)
    #Create an empty dictionary to create tree    
    if tree is None:                    
        tree={}
        tree[node] = {}
    
    if node in cat_list:
        attValue = np.unique(df[node]) #[overcast, sunny, rain]
        print(attValue)
        #We make loop to construct a tree by calling this function recursively. 
        #In this we check if the subset is pure and stops if it is pure. 
        for value in attValue:
            subtable = get_subtable(df,node,value)
            clValue,counts = np.unique(subtable['left'],return_counts=True)
            # clValue,counts = np.unique(subtable['play'],return_counts=True)   #clValue = ['no', 'yes'] <- array of 
                                                                              #unique values , counts = [3,2]<- means 
                                                                             #3 no's and 2 yes's

            if len(counts)==1:  #Checking purity of subset
                tree[node][value] = clValue[0] 
            else:        
                tree[node][value] = buildTree(subtable, height+1) #Calling the function recursively 
    elif node in num_list:
        lessEq_df, more_df = get_num_subtable(df, node, split)
#         print("LESS")
#         print(lessEq_df)
#         print("MORE")
#         print(more_df)
        clValue,counts = np.unique(lessEq_df['left'],return_counts=True)
        if len(counts)==1:  #Checking purity of subset
            tree[node][split] = clValue[0] 
        else:        
            tree[node][split] = buildTree(lessEq_df, height+1) #Calling the function recursively 
        clValue,counts = np.unique(more_df['left'],return_counts=True)
        if len(counts)==1:  #Checking purity of subset
            tree[node][float(split) + 0.000002] = clValue[0] 
        else:        
            tree[node][float(split) + 0.000002] = buildTree(more_df, height+1) #Calling the function recursively 
        
       
    return tree


```
