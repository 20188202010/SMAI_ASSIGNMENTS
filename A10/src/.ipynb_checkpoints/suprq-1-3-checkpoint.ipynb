### HMM vs RNN

#### 1) HMMs are simpler models than RNNs. So in terms of complexities, we can choose the simpler model when we have little data, and go for the more complex model when you have more data.
#### 2) Related to the previous point, HMMs make the Markovian assumption - that is, we assume that the current state depends only on the previous state. So if our data satisfy that assumption roughly, HMMs might be the preferred method, because RNNs might find spurious patterns, and hence overfit. This can easily be seen with the graph below. If your data do not satisfy that assumption, we should go with RNNs.

#### 3) With HMMs, the inputs and the outputs have a one-to-one correspondence — for instance, in part-of-speech tagging, each word in the input is mapped to a single POS. So HMMs might be more suitable here, compared to a task like translating from one language to another, where several words in one language could map to a single word in the other language, in which case RNNs should be used.
#### 4) The HMMs were better in capturing strong discriminative features
#### 5) HMM is used for early learning (short-term action memory) or when quick reaction towards new stimuli is required and LSTM when        enough data is available (long-term)
#### 6) For HMMs, we are doing a lot of the modelling based on knowledge of the problem domain in order to restrict or constrain the structure of the probabilistic (Markov chain) models. We use knowledge of the measurement process to construct sensible probabilistic models for the HMM.
#### 7) HMMs are generative models, while RNNs are primarily discriminative models. So if your task requires modelling how the data were generated, then RNNs can't be used.
